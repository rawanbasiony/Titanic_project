```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt
sns.set()
%matplotlib inline
```

```python
df = pd.read_csv('titanic (1).csv')
```

```python
df.drop(columns='Cabin',inplace=True)
```

```python
df
```

```python
df["Age"]=df["Age"].replace(np.nan,df["Age"].mean() )
df.sample(frac=0.30)
```

```python
df["Age"].values
```

```python
df['Age'] = df['Age'].astype(int)
df["Age"].values
```

```python
df['Embarked'].values
```

```python
df['Embarked']=df['Embarked'].replace(np.nan,'S')
```

```python
df['Embarked'].values
```

```python
df.sample(frac=0.20)
```

```python
df["Age"].value_counts()
```

```python
def age_group(age):
    if age > 40:
        return 'Over 40'
    elif age > 40 & age < 20:
        return '20-40'
    else:
        return '20 or less'

df['Age_group'] = df['Age'].apply(age_group)
df['Age_group'].value_counts() 
```

```python
plt.figure(figsize=(6, 4))
sns.histplot(df['Age'], kde=True, color='lightcoral', bins=30)
plt.title('Distribution of Age')
plt.show()
```

```python
df.groupby('Age_group')['Survived'].mean()
```

```python
sns.kdeplot(df[df['Survived'] == 1]['Age'], label='Survived', fill=True)
sns.kdeplot(df[df['Survived'] == 0]['Age'], label='Not Survived', fill=True)
plt.title("Age Distribution by Survival")
plt.show()
```

```python
df['Embarked'].value_counts()
```

```python
plt.figure(figsize=(6, 4))
sns.histplot(df['Embarked'], kde=True, color='blue', bins=30)
plt.title('Distribution of Embarked')
plt.show()
```

```python
df.groupby('Embarked')['Survived'].mean()
```

```python
sns.countplot(x='Embarked', data=df, hue='Survived')
plt.title("Embarked Port vs Survival")
plt.show()
```

```python
df['Fare'].value_counts()
```

```python
df['Fare'] = df['Fare'].round()
df['Fare'].value_counts()
```

```python
def fare_group(fare):
    if fare > 100:
        return 'Over 100'
    elif fare > 50:
        return '51-100'
    else:
        return '50 or less'

df['Fare_group'] = df['Fare'].apply(fare_group)
df['Fare_group'].value_counts() 
```

```python
plt.figure(figsize=(6, 4))
sns.histplot(df['Fare_group'], kde=True, color='green', bins=30)
plt.title('Distribution of Fare')
plt.show()
```

```python
df.groupby('Fare_group')['Survived'].mean()
```

```python
sns.kdeplot(df[df['Survived'] == 1]['Fare'], label='Survived', fill=True)
sns.kdeplot(df[df['Survived'] == 0]['Fare'], label='Not Survived', fill=True)
plt.title("Fare Distribution by Survival")
plt.show()
```

```python
total_pclass_1 = df[df["Pclass"] == 1].shape[0]
survived_pclass_1 = df[(df["Pclass"] == 1) & (df["Survived"] == 1)].shape[0]
survival_rate_pclass_1 = survived_pclass_1 / total_pclass_1 * 100
survival_rate_pclass_1
```

```python
total_pclass_2 = df[df["Pclass"] == 2].shape[0]
survived_pclass_2 = df[(df["Pclass"] == 2) & (df["Survived"] == 1)].shape[0]
survival_rate_pclass_2 = survived_pclass_2 / total_pclass_2 * 100
survival_rate_pclass_2
```

```python
total_pclass_3 = df[df["Pclass"] == 3].shape[0]
survived_pclass_3 = df[(df["Pclass"] == 3) & (df["Survived"] == 1)].shape[0]
survival_rate_pclass_3 = survived_pclass_3 / total_pclass_3 * 100
survival_rate_pclass_3
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='Pclass', data=df, color='skyblue' )
plt.title('Passenger Class (Pclass) Distribution')
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title("Survival Count by Passenger Class (Pclass)")
plt.xlabel("Passenger Class")
plt.ylabel("Count")
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.histplot(df['Sex'], kde=True, color='red', bins=30)
plt.title('Distribution of Sex')
plt.show()
```

```python
total_sex_male = df[df['Sex'] == 'male'].shape[0]
survived_sex_male = df[(df["Sex"] == "male") & (df["Survived"] == 1)].shape[0]
survival_rate_sex_male = survived_sex_male / total_sex_male *100
survival_rate_sex_male
```

```python
total_sex_female = df[df['Sex'] == 'female'].shape[0]
survived_sex_female = df[(df["Sex"] == "female") & (df["Survived"] == 1)].shape[0]
survival_rate_sex_female = survived_sex_female / total_sex_female *100
survival_rate_sex_female
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title("Survival Count by Passenger Class (Pclass)")
plt.xlabel("Passenger Class")
plt.ylabel("Count")
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```
